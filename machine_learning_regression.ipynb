```python
df_encoded = pd.get_dummies(df,columns=["mainroad","guestroom","basement","hotwaterheating","airconditioning","prefarea","furnishingstatus"])
df_encoded
```

```python
df_encoded.columns
```

```python
x=df_encoded [["area",	"bedrooms",	"bathrooms",	"stories",
               "parking",	"mainroad_no",	"mainroad_yes",	"guestroom_no",	"guestroom_yes",
               "basement_yes",	"hotwaterheating_no",	"hotwaterheating_yes",	"airconditioning_no",
               "airconditioning_yes",	"prefarea_no",	"prefarea_yes",	"furnishingstatus_furnished",
               "furnishingstatus_semi-furnished",	"furnishingstatus_unfurnished"]]
y=df_encoded["price"]
```

```python
import matplotlib.pyplot as plt
```

```python
!pip install -U scikit-learn
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
x_train, x_test,y_train, y_test = train_test_split(x,y, train_size=0.2, random_state=42)
```

```python
model = LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
plt.scatter(y_test,y_test, color = "black", label = "Actual")

plt.scatter(y_test,y_pred, color = "blue", label = "Predict")
plt.legend()
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
np.random.seed(0)
hours_study = np.random.uniform(0,10,100)
exam_score = 50 + 5*hours_study + np.random.normal(0,2,100)
```

```python
data = pd.DataFrame({'Hours_study': hours_study, 'Exam_score': exam_score})
data
```

```python
a = data[['Hours_study']]
b = data[['Exam_score']]
```

```python
a_train, a_test,b_train, b_test = train_test_split(a,b, train_size=0.2, random_state=42)
```

```python
model = LinearRegression()
model.fit(a_train,b_train)
```

```python
b_pred = model.predict(a_test)
b_pred
```

```python
mse = mean_squared_error(b_test, b_pred)
mse
```

```python
r2 = r2_score(b_test, b_pred)
r2
```
